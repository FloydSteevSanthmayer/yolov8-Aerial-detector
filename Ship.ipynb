```python
def train_model(config, data_file, epochs, batch_size, verbose, seed, validate, save_dir):
    """Train a YOLO model from scratch."""
    model = YOLO(config)
    model.train(
        data=data_file,
        epochs=epochs,
        batch=batch_size,
        verbose=verbose,
        seed=seed,
        val=validate,
        save_dir=save_dir,
        pretrained= "yolov8n.pt"
    )
    print("Training completed.")
    validation_metrics = model.val()
    print("Validation Metrics:", validation_metrics)
```

```python
# Constants for paths and configurations
TRAIN_MODEL_CONFIG = "/content/dataset/yolov8n.yaml"
DATA_FILE = "/content/dataset/ships.yaml"
EPOCHS = 100 # avoid overfitting
BATCH_SIZE = 4
VERBOSE = True
SEED = 0
VALIDATE = True
SAVE_DIR = "/content/dataset/"
```

```python
train_model(TRAIN_MODEL_CONFIG, DATA_FILE, EPOCHS, BATCH_SIZE, VERBOSE, SEED, VALIDATE, SAVE_DIR)
```

```python
from ultralytics import YOLO
import glob
import math
import cv2

MODEL_PATH = "/content/runs/detect/train/weights/best.pt"
IMAGES_PATH = "/content/dataset/ship/test/images/*"
SAVE_PATH = "/content/test_images/"

model = YOLO(MODEL_PATH)

classNames = ["ship"]
images=glob.glob(IMAGES_PATH)
confidence = 0.01

for image in images:
    frame = cv2.imread(image)  # Read image from file

    # Perform object detection on the frame
    results = model(frame)

    # Loop through the detected objects in the frame

    for r in results:
        boxes = r.boxes
        for box in boxes:
            # Calculate confidence level
            _confidence = math.ceil((box.conf[0] * 100)) / 100
            print("Confidence --->", _confidence)

            # Skip if confidence is below threshold
            if _confidence < confidence:
                continue

            # Extract bounding box coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Convert to integer values

            # Draw bounding box rectangle
            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 255), 3)

            # Get class label
            cls = int(box.cls[0])

            # Prepare text for class label
            org = [x1, y1]
            font = cv2.FONT_HERSHEY_SIMPLEX
            fontScale = 1
            color = (255, 0, 0)
            thickness = 2

            # Draw class label on the frame
            cv2.putText(frame, classNames[cls], org, font, fontScale, color, thickness)
    image_name = image.split("/")[-1]
    cv2.imwrite(SAVE_PATH+image_name, frame)
```
